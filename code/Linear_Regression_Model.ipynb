```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('dataset1.csv')
df.head()
df.info()

# Select Data Used for correlation analysis
data = df[['Illness_code','Visitors with Patient','Admission_Deposit','Age Value','LOS']]
data.head()
```

We just select 4 variables as X variables and then we make the EDA

```python
# EDA
print(data.isnull().sum())
print(data.describe().round(2))

# Set the figure size
plt.figure(figsize=(15, 10))

# Create the distribution plot
sns.distplot(data['Admission_Deposit'])

# Set chart title
plt.title('Distribution of Admission Deposit', fontsize=14)

# Set x-axis label
plt.xlabel('Admission Deposit', fontsize=12)

# Set y-axis label
plt.ylabel('Density', fontsize=12)

# Show the plot
plt.show()
```

```python
# Set the figure size
plt.figure(figsize=(15, 10))

# Create the distribution plot
sns.distplot(data['Visitors with Patient'], color='orange')

# Set chart title
plt.title('Distribution of Visitors with Patient', fontsize=14)

# Set x-axis label
plt.xlabel('Visitors with Patient', fontsize=12)

# Set y-axis label
plt.ylabel('Density', fontsize=12)

# Show the plot
plt.show()
```

```python
# Set the figure size
plt.figure(figsize=(15, 10))

# Create the count plot
sns.countplot(x=data['Illness_code'].sort_values(), data=data, palette='crest')

# Set the axis labels
plt.xlabel('Severity of Illness', fontsize=12)
plt.ylabel('Count', fontsize=12)

# Set the title
plt.title('Distribution of Severity of Illness', fontsize=14)

# Show the plot
plt.show()
```

```python
# Set the figure size
plt.figure(figsize=(15, 10))

# Create the distribution plot
sns.distplot(data['Age Value'])

# Set chart title
plt.title('Distribution of Age', fontsize=14)

# Set x-axis label
plt.xlabel('Age', fontsize=12)

# Set y-axis label
plt.ylabel('Density', fontsize=12)

# Show the plot
plt.show()
```

```python
# Set the figure size
plt.figure(figsize=(15, 10))

# Create the distribution plot
sns.distplot(data['LOS'])

# Set chart title
plt.title('Distribution of Length of Stay', fontsize=14)

# Set x-axis label
plt.xlabel('Length of Stay', fontsize=12)

# Set y-axis label
plt.ylabel('Density', fontsize=12)

# Show the plot
plt.show()
```

After EDA, we have known what is the data like. Then we will make the correaltion analysis

```python
# Liner Regression
import statsmodels.api as sm

# Define X and y
X = data.drop('LOS', axis=1)
y = data['LOS']

# Add intercept
X = sm.add_constant(X)

# Fitting the model
model = sm.OLS(y, X).fit()

# Output model summary
print(model.summary())
```
